```python
import matplotlib.pyplot as plt
import numpy as np
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]
plt.ylim(0,50)
plt.bar(categories,data)
plt.grid(True, axis = 'both')
plt.xticks(rotation = 45)
plt.title('Bar Chart', loc = 'center', pad = 20)
plt.xlabel('Categories')
plt.ylabel('values')

plt.show()
```

```python
labels = ['Apple', 'Banana', 'Mango', 'Blueberry']
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # Banana 부분에서만 강조하기 위해
plt.pie(sizes,labels = labels,autopct = '%.1f%%',startangle = 90, explode = explode, shadow = True) # = 안쓰려면 ratio, explode,label autopct,shadow 순으로
plt.show()
```

```python
x = np.linspace(0, 10, 100)
a = np.sin(x)
b = np.cos(x)
plt.plot(x,a, color = 'blue', label = 'sin(x)') #plt.plot(x,a,x,b) 로 하면 한번에 그리기 가능
plt.plot(x,b, color = 'red', label = 'cos(x)')
plt.legend() #plt.legend(loc = 'lower left', ncol = 1, fontsize = 10, frameon = True)결과값 같음
plt.title('Sine and Cosine Functions', loc = 'center', pad = 20)
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True, axis = 'both')
plt.show()
```
